```python
df = df.drop(columns=['Date'], axis=1)
df = df.rename(columns={"NDate": "Date"})

df
                    
```

```python
for i in df["Name"]:
    if (" L " or " M " or " H ") in i:
        print(True,"---------------", i)
    else:
        pass
```

```python
xxx = []
for i in df["Name"]:
    if ("Lower" in i) or ("Middle" in i) or ("Upper" in i):
        pass
    else:
        xxx.append(i)
len(xxx)
```

```python
def floor_change(x):
    if "Lower" in x:
        return "L"
    elif "Middle" in x:
        return "M"
    elif "Upper" in x:
        return "H"
    else:
        return ""
```

```python
df["Floor"] = df["Name"].apply(lambda x: floor_change(x))
```

```python
df["Floor"]
```

```python
for i in ['SA (ft2)', 'Price(HK$)']:
    df[i] = df[i].str.replace(r'\D', '')
    df[i] = df[i].astype(int)
    
```

```python
df.dtypes
```

```python
df["Price/ft2(SA)"] = df["Price(HK$)"]/df["SA (ft2)"]
df = df.round({'Price/ft2(SA)': 2})

df.dtypes
```

```python
index_names = df[(df['Floor'] == "")|(df['Date'] >= "2022-05-01")].index

# drop these row indexes
# from dataFrame
df.drop(index_names, inplace = True)
```
