```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (12, 12)
```

```python
f = ["H","M","L"]

def fc(x):
    if x == "H":
        return "Upper"
    if x == "M":
        return "Middle"
    if x == "L":
        return "Lower"

for i in f:
    temp_df = pd.DataFrame(df[df['Floor'] == i])
    temp_df.groupby('District1')["Price/ft2(SA)"].mean().sort_values().plot(kind="barh",figsize=(4,6))
    plt.title("Avg. rent/ft2 by district ({} Floor)".format(fc(i)))
    plt.xticks(rotation=0)
    plt.xlabel('Rent/ft2',size = 10)
    plt.show()
```
